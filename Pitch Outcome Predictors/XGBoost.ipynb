```python
import numpy as np
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
```

```python
# Load preprocessed data
flat_X_train = np.load('flat_X_train.npy')
y_train = np.load('y_train.npy')

flat_X_val = np.load('flat_X_val.npy')
y_val = np.load('y_val.npy')

flat_X_test = np.load('flat_X_test.npy')
y_test = np.load('y_test.npy')

print(flat_X_train.shape, y_train.shape)
print(flat_X_val.shape, y_val.shape)
print(flat_X_test.shape, y_test.shape)

pitch_sequence_length, num_classes, flat_num_features = np.load('flat_data_parameters.npy')

print(pitch_sequence_length, num_classes, flat_num_features)
```

```python
# Initialize the model
xgb_model = xgb.XGBClassifier(objective='multi:softmax', num_class=num_classes, eval_metric='mlogloss')

# Train the model
xgb_model.fit(flat_X_train, y_train)

# Predict on the test set
y_test_pred = xgb_model.predict(flat_X_test)
```

```python
np.unique(y_test_pred) 
```

```python
f1_score(y_test, y_test_pred, average='macro')
```
