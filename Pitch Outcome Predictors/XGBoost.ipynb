```python
import numpy as np
import xgboost as xgb
from xgboost import XGBClassifier

from sklearn.metrics import f1_score
```

```python
# Load preprocessed data
X_train = np.load('X_train.npy')
y_train = np.load('y_train.npy')

X_val = np.load('X_val.npy')
y_val = np.load('y_val.npy')

X_test = np.load('X_test.npy')
y_test = np.load('y_test.npy')

print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)
print(X_test.shape, y_test.shape)

pitch_sequence_length, num_classes, num_features = np.load('data_parameters.npy')

print(pitch_sequence_length, num_classes, num_features)
```

```python
total_num_features = (pitch_sequence_length - 1) * num_features
print(total_num_features)
```

Flatten X into a 2D array of shape (number of sequences, total_num_features) because we do not care about the time series aspect of the data.

```python
flat_X_train=X_train.reshape(X_train.shape[0], -1)
flat_X_val=X_val.reshape(X_val.shape[0], -1)
flat_X_test=X_test.reshape(X_test.shape[0], -1)

print(flat_X_train.shape, flat_X_val.shape, flat_X_test.shape)
```

```python
# Initialize the model
xgb_model = xgb.XGBClassifier(objective='multi:softmax', num_class=num_classes, eval_metric='mlogloss')

# Train the model
xgb_model.fit(flat_X_train, y_train)

# Predict on the test set
y_pred = xgb_model.predict(flat_X_test)
```

```python
np.unique(y_pred) 
```

```python
f1_score(y_test, y_pred, average='macro')
```
